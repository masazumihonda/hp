## For $\theta =0$, explicit form of the Hamiltonian  is
$$ H = H_Z +H_{ZZ}  +H_{XX} +H_{YY} $$,
$$ H_Z =\frac{m}{2}\sum_{n=1}^N (-1)^n Z_n 
         -\frac{J}{2}\sum_{n=1}^{N-1} (n\ {\rm mod}\ 2) \sum_{\ell=1}^n Z_\ell $$
$$H_{ZZ}= \frac{J}{2} \sum_{n=2}^{N-1} \sum_{1\leq k<\ell \leq n} Z_k Z_\ell $$         
$$ H_{XX} = \frac{w}{2} \sum_{n=1}^{N-1} X_n X_{n+1} $$
$$ H_{YY} = \frac{w}{2} \sum_{n=1}^{N-1} Y_n Y_{n+1} $$


### The ground state in the large mass limit is
$$ |{\rm massive} \rangle \equiv |0101\cdots \rangle = |0\rangle \otimes X|0\rangle \cdots \otimes |0\rangle \otimes X|0\rangle  $$
### We are intested in computing the quantity
$$ P(t) \equiv \left| \langle{\rm massive} | e^{-iHt} |{\rm massive} \rangle \right|^2 $$
